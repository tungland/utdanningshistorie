```python
import dhlab as dh
```

```python
lærebøker = dh.Corpus(doctype="digibok", subject="lærebøker", limit=10000000)
```

```python
norsk = dh.Corpus(doctype='digibok', subject="lærebøker AND norsk", limit=10000000)
```

```python
norsk.frame.groupby('year').dhlabid.count().plot(kind='bar', figsize=(20, 15))
```

```python
lærebøker.frame.groupby("year").dhlabid.count().sort_index()
```

```python
lærebøker.frame.groupby("year").dhlabid.count().sort_index()
```

```python
lærebøker.frame.groupby("year").dhlabid.count().sort_index().plot(kind='bar', figsize=(30, 20))
```

```python
df = norsk.frame.copy().groupby("year").dhlabid.count().to_frame("books").reset_index()
```

```python
df
```

```python
df_repeated = df.loc[df.index.repeat(df['books'])].reset_index(drop=True)
```

```python
mean_year = df_repeated['year'].mean()
median_year = df_repeated['year'].median()

print("Mean Year:", mean_year)
print("Median Year:", median_year)
```
