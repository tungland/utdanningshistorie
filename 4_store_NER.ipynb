```python
import dhlab as dh
from dhlab.text.parse import NER, Models
```

```python
m = Models()
m.models
```

```python
norsk = dh.Corpus(doctype='digibok', subject="lærebøker AND norsk", limit=10000000)
```

```python
df = norsk.frame.drop_duplicates(subset='urn')
```

```python
df.loc[norsk.frame['urn'] == "URN:NBN:no-nb_digibok_2012112012001"]
```

```python
result = NER(urn="URN:NBN:no-nb_digibok_2012112012001", model='da_core_news_lg', start_page=0, to_page=1000000)
```
